```python
#plotting correlations
num_feat=houses.columns[houses.dtypes!=object]
num_feat=num_feat[1:-1]
labels=[]
values=[]
for col in num_feat:
    labels.append(col)
    values.append(np.corrcoef(houses[col].values, houses.SalePrice.values)[0,1])
    
    ind=np.arange(len(labels))
    width=0.9
    fig, ax=plt.subplots(figsize=(12,40))
    rects=ax.barh(ind, np.array(values), color='red')
    ax.set_yticks(ind+((width)/2.))
    ax.set_yticklabels(labels,rotation='horizontal')
    ax.set_xlabel("Correlation coefficient")
    ax.set_title("Correlation Coefficients w.r.t Sale Price");
    
```

```python
##MIssing value imputation
#Missing values in the trainig data set can affect prediction or classification of a model negatively.
#Also, some machine learning algorithms can't accept missing data such as SVM.
#But filling missing values with mean/median/mode or using another predictive model is also risky, which brings out not a certain figure.
#Therefore, we can use models like Decision Trees and Random Forest which handel missing values very well.

```

```python
#Let's see if there are any colums with missing values.
null_columns=houses.columns[houses.isnull().any()]
houses[null_columns].isnull().sum()
```

```python
labels=[]
values=[]
for col in null_columns:
    labels.append(col)
    values.append(houses[col].isnull().sum())
ind = np.arange(len(labels))
width=0.9
fig, ax=plt.subplots(figsize=(12,50))
rects=ax.barh(ind, np.array(values),color='violet')
ax.set_yticks(ind+((width)/2.))
ax.set_yticklabels(labels, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_ylabel("Column Names")
ax.set_title("Variables with missing values")
```
